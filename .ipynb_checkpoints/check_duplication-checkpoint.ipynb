# Check review duplication 

<h3> Import data:

```python
import pandas as pd
import glob

files = glob.glob('./CSV_files/*.csv')
df = pd.concat([pd.read_csv(f,header=None) for f in glob.glob('./CSV_files/*.csv')])
df = df.rename(columns = {0:'reviews'})
df
```

<h3> check duplicated reviews:

```python
lst = []
dup_count = 0
for review in df['reviews']:
    if review not in lst:
        lst.append(review)
        
    else:
        dup_count+=1
        print('duplicated review:{}'.format(review))

print('duplicated review counts: %d' % dup_count)
```

duplicated review counts: 229
